## <span style='color:orange'>Fixing NaN Values</span>

Below columns are having maximum number of null values in their entries compared to other columns. But these columns should be numerical and we are considering NaN in the excel means there is no case registered for this age group. Hence all NaN for these columns should be replaced with zero(0).

1. **transgender_below_12_years**
2. **transgender_12_years_to_16_years**
3. **transgender_16_years_to_18_years**
4. **transgender_children** and
5. **transgender_18_years_and_above**

```python
len(mpd["transgender_below_12_years"])
mpd['transgender_below_12_years'].isnull().sum()
mpd['transgender_below_12_years'] = mpd['transgender_below_12_years'].fillna(0)
mpd['transgender_below_12_years'].isnull().sum()
len(mpd["transgender_12_years_to_16_years"])
mpd["transgender_12_years_to_16_years"].isnull().sum()
mpd["transgender_12_years_to_16_years"] = mpd["transgender_12_years_to_16_years"].fillna(0)
mpd["transgender_12_years_to_16_years"].isnull().sum()
print(len(mpd["transgender_16_years_to_18_years"]))
mpd["transgender_16_years_to_18_years"].isnull().sum()
mpd["transgender_16_years_to_18_years"] = mpd["transgender_16_years_to_18_years"].fillna(0)
mpd["transgender_16_years_to_18_years"].isnull().sum()
print(len(mpd["transgender_children"]))
mpd["transgender_children"].isnull().sum()
mpd["transgender_children"] = mpd["transgender_children"].fillna(0)
mpd["transgender_children"].isnull().sum()
print(len(mpd["transgender_18_years_and_above"]))
mpd['transgender_18_years_and_above'].isnull().sum()
mpd['transgender_18_years_and_above'] = mpd['transgender_18_years_and_above'].fillna(0)
mpd['transgender_18_years_and_above'].isnull().sum()
cmpd = mpd[['transgender_12_years_to_16_years', 'transgender_16_years_to_18_years', 'transgender_children', 'transgender_18_years_and_above']]
cmpd
```

#### <span style='color:red'>Box Plots</span>

```python
# Select relevant columns for the box plot
boxplot_data = mpd[['male_children', 'female_children', 'transgender_children',
                   'male_18_years_and_above', 'female_18_years_and_above', 'transgender_18_years_and_above']]

# Reshape the data for seaborn
melted_data = boxplot_data.melt(var_name='Category', value_name='Count')

# Create the box plot
plt.figure(figsize=(12, 6))
ax=sns.boxplot(x='Category', y='Count', data=melted_data)
plt.title('Box Plot of Missing Persons by Category')
plt.xticks(rotation=45)
plt.grid(True)
# Get box plot data
box_data = [box.get_ydata() for box in ax.artists]

# Iterate through box plots and add text annotations
for i in range(len(boxplot_data.columns)):  # Adjusted loop
    # Get the median, Q1, Q3 from the statistical data
    median = melted_data[melted_data['Category'] == boxplot_data.columns[i]]['Count'].median()
    q1 = melted_data[melted_data['Category'] == boxplot_data.columns[i]]['Count'].quantile(0.25)
    q3 = melted_data[melted_data['Category'] == boxplot_data.columns[i]]['Count'].quantile(0.75)

    # Define an offset for the y-coordinate
    x_offset = 0.1
    # Add text annotations for median, Q1, and Q3
    ax.text(i+x_offset, q1+ 400, f'Q1:{q1:.2f}', ha='left', va='center', color='blue')
    ax.text(i+x_offset, median+ 800,f'Median: {median:.2f}', ha='left', va='center', color='black')
    ax.text(i+x_offset, q3+ 1200, f'Q3:{q3:.2f}', ha='left', va='center', color='red')

plt.tight_layout()
plt.show()
```

The above box plot provides a statistical summary of missing persons by various categories (e.g., male children, female children, transgender children, etc.). It shows the distribution of missing persons for each category, indicating the median, quartiles, and potential outliers in the data.

- The median, first quartile (Q1), and third quartile (Q3) values are annotated for each category(male_children, female_childre.
- The boxes represent the interquartile range (IQR), and the whiskers extend to the smallest and largest values within 1.5 times the IQR from the first and third quartiles, respectively.
- Outliers, if any, are represented as points beyond the whiskers.

This plot helps identify the spread and central tendency of missing persons data for each category, and highlights the presence of any significant outliers.

#### <span style='color:red'>Lambda Functions â€“ Simple Data Transformation using Lambda</span>

```python
# Categorize total missing persons into bins: Low, Medium, High
mpd['missing_severity'] = mpd['total_missing'].apply(
    lambda x: 'Low' if x < 50 else ('Medium' if x < 150 else 'High')
)
print("Missing Severity:\n",mpd['missing_severity'].head())
```

```python
# Flag districts with high female missing count
mpd['high_female_missing'] = mpd['total_females'].apply(lambda x: True if x > 100 else False)
print("High Female Missing Flag:\n", mpd[['total_females', 'high_female_missing']].head())
```

```python
# Tagging even/odd years
mpd['year_type'] = mpd['year'].apply(lambda x: 'Even' if int(x) % 2 == 0 else 'Odd')
print("Year Type Odd:\n", mpd[['year', 'year_type']].head(),"\nYear Type Even:\n",mpd[['year', 'year_type']].tail())
```

#### <span style='color:red'>User-Defined Functions</span>

```python
# Classify male missing count
def classify_male_missing(count):
    if count < 30:
        return 'Low'
    elif count < 100:
        return 'Medium'
    else:
        return 'High'

mpd['male_missing_class'] = mpd['total_males'].apply(classify_male_missing)
print("Male Missing Classification:\n", mpd[['total_males', 'male_missing_class']].head())
```

```python
# Combine gender data into a string

# total TG
mpd['total_transgender'] = mpd['transgender_children'] + mpd['transgender_18_years_and_above']

def summary_string(row):
    return f"Male: {row['total_males']}, Female: {row['total_females']}, TG: {row['total_transgender']}"

mpd['gender_summary'] = mpd.apply(summary_string, axis=1)

print("Gender Summary:\n", mpd['gender_summary'].head())
```

```python
# District flag based on total count

def district_level(total):
    return 'Severe' if total > 200 else 'Normal'

mpd['district_flag'] = mpd['total_missing'].apply(district_level)
print("District Severity Flag:\n", mpd[['total_missing', 'district_name','district_flag']].head())
```
